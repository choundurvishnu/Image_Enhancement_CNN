**Histogram**

```python
import cv2
import matplotlib.pyplot as plt
from IPython.display import Image

img_path = '/content/drive/MyDrive/data/test_data/DIC/DICM/01.jpg'
# Display the image
Image(filename=img_path)
image = cv2.imread(img_path)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])


# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Display the image on the left subplot
ax1.set_title('Image')
ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax1.axis('off')

# Display the histogram on the right subplot
ax2.set_title('Histogram')
ax2.plot(histogram, color='k')
ax2.set_xlim([0, 256])
ax2.set_xlabel('Pixel Value')
ax2.set_ylabel('Frequency')

# Show the figure with image and histogram side by side
plt.tight_layout()
plt.show()
```

RGB Histogram

```python
import cv2
import matplotlib.pyplot as plt
img_path = '/content/drive/MyDrive/data/test_data/DIC/DICM/01.jpg'
# Display the image
Image(filename=img_path)
image = cv2.imread(img_path)

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Display the image on the left subplot
ax1.set_title('Image')
ax1.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
ax1.axis('off')

# Calculate and display the RGB histograms on the right subplot
ax2.set_title('RGB Histograms')
for i, col in enumerate(['b', 'g', 'r']):
    hist = cv2.calcHist([image], [i], None, [256], [0, 256])
    ax2.plot(hist, color=col, label=col.upper())
ax2.set_xlim([0, 256])
ax2.set_xlabel('Pixel Value')
ax2.set_ylabel('Frequency')
ax2.legend()

# Show the figure with image and histogram side by side
plt.tight_layout()
plt.show()
```
